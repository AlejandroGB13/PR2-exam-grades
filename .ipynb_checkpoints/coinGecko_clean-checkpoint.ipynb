No es una novedad el creciente interés por la compra-venta de monedas virtuales. Con el programa resultante y conocimientos de cómo fluctúa el mercado sería posible encontrar los momentos idóneos para realizar las transacciones y maximizar los beneficios.

### Importar librerías

```python
import pandas as pd
```

### Carga de datos

```python
df = pd.read_csv('./CoinGecko_reviewed.csv', sep=';')
print(df)
```

# 2. Limpieza de datos

El dataframe contiene 7000 registros y 7 variables. 

Contiene 100 tipos de monedas y sus estadísticas durante un intervalo de una hora.

Las variables son: Name, Time, Price, Min_price_24h, Max_price_24h, Market_cap, Vol_24h.

## 2.1. Selección de variables

Al ser un conjunto de datos formado por nosotros mismos con los valores de CoinGecko, todas las variables ya son las que nos interesan para el análisis, asi que no descartaremos ninguna.

## 2.2. Clase de las variables

```python
df.dtypes
```

Después de guardar en csv, todas las variables son de tipo 'Object', lo que describe cómo deben interpretarse los bytes en el bloque de memoria de tamaño fijo correspondiente a un elemento de matriz.

Pero realmente las clases son las siguientes:


```python
num_cols = ['Price', 'Min_price_24h', 'Max_price_24h', 'Market_cap', 'Vol_24h']
df[num_cols] = df[num_cols].astype(str)

df[num_cols] = df[num_cols].replace({'\$': '', ',': ''}, regex=True)
df[num_cols] = df[num_cols].astype(float)
df.dtypes
```

## 2.3. Análisis de los outliers

En nuestro conjunto de datos tenemos los valores de distintas monedas que difieren mucho en estos valores. A diferencia de un conjunto de datos donde se analiza por ejemplo el peso de las personas, en donde el peso tiende a un rango de valores, en nuestro caso no tenemos ninguna tendencia ya que los precios pueden diferir completamente entre una moneda u otra sin relación alguna. Por lo tanto los resultados que podamos obtener no pueden ser outliers.
